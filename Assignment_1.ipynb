## Assignment 1: Calculating a stock's 5-year price & returns and plotting them on charts.
**For this assignment, I choose Tata Steel Ltd. (NSE: TATASTEEL) which is listed on the NSE (India). Data for this assignment was gathered using Google Finance.**  
  
The solutions to both the questions of the assignment have been clubbed in this single Jupyter notebook for easy reference.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("TATASTEEL_5Y_Price.csv")
```

```python
df.head()
```

```python
df[['Date','time']] = df['Date'].str.split(' ',expand=True)
```

```python
df
```

```python
df.drop('time', inplace=True, axis=1)
```

```python
df
```

```python
df.rename(columns = {'Close' : 'Price'}, inplace = True)
df
```

```python
df.set_index('Date', inplace=True)
df
```

```python
df['Return'] = df['Price'].pct_change(1)
```

```python
df
```

```python
df['Price'].plot(figsize = (12, 8))
```

```python
df['Return'].plot(figsize = (12, 8))
```
